```python
# Gán giá trị thiếu trong cột Review Rating với Review Rating trung bình của Category đó.
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# Đổi tên cột theo kiểu snake casing để dễ thao tác

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# Tạo phân khúc tuổi, để tìm hiểu hành vi, tối ưu insight
bins = [18, 25, 35, 45, 55, float('inf')]
labels = ['Gen Z', 'Young Adult', 'Adult', 'Middle Age', 'Senior']

df['age_group'] = pd.cut(
    df['age'],
    bins=bins,
    labels=labels,
    right=False
)
```

- Gen Z (18–24): Thích trend, giá rẻ, mua cảm xúc  
- Young Adult (25–34): Bắt đầu kiếm tiền, chi tiêu tăng  
- Adult (35–44): Ổn định tài chính, mua có kế hoạch  
- Middle Age (45–54): Ưu tiên chất lượng, gia đình  
- Senior (55+): Ít mua hơn, trung thành cao  

```python
# Kiểm tra các giá trị gán
df[['age','age_group']].head(10)
```

```python
# tạo cột purchase_frequency_days mới, chuyển từ text -> number

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# Kiểm tra liệu 2 cột này có giá trị toàn giống nhau không, để loại bỏ cột dư thừa.
df[['discount_applied','promo_code_used']].head(10)
```

```python
# Kiểm tra
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Loại bỏ cột dư thừa.
df = df.drop('promo_code_used', axis=1)
```
